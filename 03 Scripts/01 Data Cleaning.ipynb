### 3. Data Consistency Checks: cwur_data.csv

```python
# Check dataframe dimensions
df_cwur.shape
```

```python
# Check the descriptive statistics
df_cwur.describe()
```

```python
# Check for mixed datatypes
for col in df_cwur.columns.tolist():
  weird = (df_cwur[[col]].applymap(type) != df_cwur[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_cwur[weird]) > 0:
    print (col, ' mixed')
  else: print(col, ' consistent')
```

```python
# Check the datatypes of each column
df_cwur.dtypes
```

```python
# Check for missing values
df_cwur.isnull().sum()
```

> The column 'broad_impact' contains missing values. These missing values will remain as NaN.

```python
# Check for duplicates
df_cwur_dups = df_cwur[df_cwur.duplicated()]
df_cwur_dups.size
```

```python
# Export checked dataframe
df_cwur.to_csv(os.path.join(path, '02 Data','Prepared Data', 'cwur_checked.csv'))
```
